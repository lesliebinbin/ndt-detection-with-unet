```python
from pathlib import Path
```

```python
src_folder = Path("origin_jpg_data")
```

```python
files01 = [file.name for file in Path("dataset_by_aspect_ratio/train/circle").iterdir()]
files02 = [file.name for file in Path("dataset_by_aspect_ratio/valid/circle").iterdir()]
from PIL import Image
current_files = []
for file in src_folder.rglob("**/*.jpg"):
    if "圆形" in str(file.absolute()):
        current_files.append(file)
```

```python
files01, files02
```

```python
general_files = [*files01, *files02]
```

```python
len(current_files), len(general_files)
```

```python
rest_files = [file for file in current_files if file.name not in general_files]
```

```python
for file in rest_files:
    img = Image.open(file)
    width, height = img.width, img.height
    if height> width:
        img.transpose(Image.Transpose.ROTATE_90).save(file)
rest_files
```

```python
from collections import defaultdict
```

```python
proportions = defaultdict(list)
for file in rest_files:
    img = Image.open(file)
    proportion = img.width // img.height
    proportions[proportion].append(file)
```

```python
proportions
```

```python
import shutil
```

```python
from sklearn.model_selection import train_test_split
```

```python
for files in proportions.values():
    if len(files) < 5:
        train_files, test_files = files, []
    else:
        train_files, test_files = train_test_split(files)
    for train_file in train_files:
        shutil.copyfile(train_file, Path("dataset_by_aspect_ratio/train/circle")/train_file.name)
    for test_file in test_files:
        shutil.copyfile(train_file, Path("dataset_by_aspect_ratio/train/circle")/test_file.name)
```

```python
import numpy as np
from PIL import Image
```

```python
from pathlib import Path
src_folder = Path("artificial_images")
tgt_folder = Path("original_images")
```

```python
import matplotlib.pyplot as plt
for file in src_folder.rglob("**/*.jpg"):
    img = np.array(Image.open(file))[:, :, 2]
    if img.shape[0] > img.shape[1]:
        img = img.T
    img = Image.fromarray(img)
    plt.axis("off")
    plt.imshow(img, cmap='gray')
    plt.show()
    tgt_path = tgt_folder / file.relative_to(src_folder)
    tgt_path.parent.mkdir(exist_ok=True, parents=True)
    img.save(tgt_path)
    break
    
```

```python
from pathlib import Path
from PIL import Image, ImageOps
```

```python
src_folder = Path("valid-versus-invalid")
src_train_folder = src_folder / "train"
src_train_enhance_folder = src_folder / "train_enhance"
```

```python
for file in src_train_folder.rglob("**/*.jpg"):
    img = Image.open(file)
    flip_lr_img = src_train_enhance_folder / file.relative_to(src_train_folder).with_name(f"{file.stem}_lr.jpg")
    flip_lr_img.parent.mkdir(exist_ok=True, parents=True)
    flip_td_img = src_train_enhance_folder / file.relative_to(src_train_folder).with_name(f"{file.stem}_td.jpg")
    flip_td_img.parent.mkdir(exist_ok=True, parents=True)
    rot_90_img = src_train_enhance_folder / file.relative_to(src_train_folder).with_name(f"{file.stem}_rot_90.jpg")
    rot_90_img.parent.mkdir(exist_ok=True, parents=True)
    print(flip_lr_img)
    print(flip_td_img)
    print(rot_90_img)
    img.transpose(Image.Transpose.FLIP_LEFT_RIGHT).save(flip_lr_img)
    img.transpose(Image.Transpose.FLIP_TOP_BOTTOM).save(flip_td_img)
    img.transpose(Image.Transpose.ROTATE_90).save(rot_90_img)
```
